# Drop The Columns ("Sno",'ConfirmedIndianNational', "ConfirmedForeignNational")

```python
cdata=cdata.drop(["Sno",'ConfirmedIndianNational', 'ConfirmedForeignNational'],axis=1)
```

```python
cdata
```

# Add  a Column of Active Cases 

```python
cdata["Active-Cases"]=cdata["Confirmed"]-cdata["Cured"]-cdata["Deaths"]
```

```python
cdata.head()
```

# Make a Pivot Table 

```python
s=pd.pivot_table(cdata,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory", aggfunc=max)
```

```python
s
```

# Add a Column Recovery Rate and Mortality Rate in pivot table

```python
s["Recovery-Rate"]=s["Cured"]*100/s["Confirmed"]
```

```python
s["Mortality-Rate"]=s["Deaths"]*100/s["Confirmed"]
```

```python
s=s.sort_values(by="Confirmed",ascending=False)
```

```python
s
```

# Most Active Case in India

```python
t=cdata.groupby("State/UnionTerritory").max()[["Active-Cases","Date"]].sort_values(by="Active-Cases",ascending=False).reset_index()
t
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Most Active case in India",size=25)
sns.barplot(data=t.iloc[:10],y="Active-Cases",x="State/UnionTerritory",linewidth=2,edgecolor="red")

```

# Top 10 Most Death Case in India

```python
y=cdata.groupby("State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by="Deaths",ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Most Deaths in India",size=25)
sns.barplot(data=y.iloc[:10],y="Deaths",x="State/UnionTerritory",linewidth=2,edgecolor="pink")
```

# Top 10 Recovered States in India

```python
r=cdata.groupby("State/UnionTerritory").max()[["Cured","Date"]].sort_values(by="Cured",ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top Recoverd State in India",size=25)
sns.barplot(data=r.iloc[:10],y="Cured",x="State/UnionTerritory",linewidth=2,edgecolor="pink")
```

# Top 5 Affected States

```python
plt.figure(figsize=(16,5))
d=cdata[cdata["State/UnionTerritory"].isin(["Maharashtra","Karnataka","Kerala","Uttar Pradesh","Tamil Nadu"])]
sns.lineplot(x="Date",y="Active-Cases",data=d,hue="State/UnionTerritory")
plt.title("Top 5 Affected State")
```

# Remove India  From  state column in vaccination Data Set

```python
vaccine=vaci[vaci["State"]!="India"]
```

```python
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
```

```python
vaccine
```

# Most Vaccinated State

```python
max_v=vaccine.groupby("State")["Total"].sum().to_frame("Total").sort_values("Total",ascending=False)
```

```python
t=max_v.head(5)
t
```

```python
fig=plt.figure(figsize=(10,5))
sns.barplot(data=t,x=t.index,y=t.Total)
plt.title("Top 5 Vaccinated State In India",size=20)
plt.show()
```

# Minimum Vaccinated States in India

```python
min_v=vaccine.groupby("State")["Total"].sum().to_frame("Total").sort_values("Total",ascending=True)
```

```python
y=min_v.head(5)
y
```

```python
fig=plt.figure(figsize=(10,5))

sns.barplot(data=y,x=y.index,y=y.Total)
plt.title("Least 5 Vaccinated State In India",size=20)
plt.xticks(rotation=35)
plt.show()
```
